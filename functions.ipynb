```python
# 3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of
# each number.

def squares(lis):
  for i in lis:
    yield i ** 2
  
lis = list(map(int, input("Enter space seperated numbers : ").split(' ')))

print(list(squares(lis)))
```

```python
# 4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
  for i in  range(2, (num//2)):
    if num % i == 0:
      return False
  else:
    return True
  
num = int(input("Enter a number inbetween 1-200"))

if num < 1 or num > 200:
  print("Invalid input.")
else:
  if is_prime(num):
    print("Prime")
  else:
    print("Not Prime")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

end_point = int(input("Enter a limit : "))

def fibbo(end_point):
  a = 0
  b = 1
  for _ in range(end_point):
    yield a
    a,b = b,a+b
  
res = fibbo(end_point)

print(next(res)) # 0
print(next(res)) # 1
print(next(res)) # 1
print(next(res)) # 2
print(next(res)) # 3
print(next(res)) # 5
print(next(res)) # 8
```

```python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

exponent = int(input("Enter the exponent limit : "))

def expo(exponent):
  for ex in range(1, exponent + 1):
    yield 2 ** ex

exponential = list(expo(exponent))

print(exponential)
```

```python
# 8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

lis = [
  (23,56,34),
  (90,46,21),
  (34,51,77)
]

# this is the way I did before knowing the concept

sorts = lambda x, y : x if x[1] < y[1] else y

import functools as ft

sorted_lis = [

]

for _ in lis:
  mins = ft.reduce(sorts, lis)
  
  sorted_lis.append(mins)

  lis.remove(mins)

sorted_lis.append(lis[0])

print(sorted_lis)

# after knowing the concept

lis = [
  (23,56,34),
  (90,46,21),
  (34,51,77)
]

lis.sort(key = lambda x : x[1])

print(lis)
```

```python
# 9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

lis = [16, 25, 35, 46] # in Celcius

temp_convert = lambda x, t='c': (x*9/5) + 32 if t == 'c' else (x-32)*5/9 # 'c' = Celcius

print(list(map(temp_convert, lis)))
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

string = "Hello, my name is smarak. I am an AIML Engineer at Cognizant!"

rem_vow = lambda x: x if x not in "aeiouAEIOU" else ''

print(list(filter(rem_vow, string)))
```
